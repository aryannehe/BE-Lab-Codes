# Problem 1: Tokenization using NLTK

**Objective:** Perform tokenization (Whitespace, Punctuation-based, Treebank, Tweet, MWE) using NLTK library. Use porter stemmer and snowball stemmer for stemming. Use any technique for lemmatization.

**Dataset:** Use any sample sentence

## 1. Import Required Libraries

```python
# Install required packages
!pip install nltk -q

import nltk
from nltk.tokenize import (
    word_tokenize,
    WhitespaceTokenizer,
    WordPunctTokenizer,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK data
nltk.download('punkt', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)

print("✓ Libraries imported successfully!")
```

## 2. Sample Sentences for Tokenization

```python
# Sample sentences for different tokenization techniques
sample_text = """Natural Language Processing (NLP) is amazing! 
It's used in chatbots, machine translation, and sentiment analysis. 
Companies like Google, Microsoft, and OpenAI are leading the field."""

# Sample for tweet tokenizer
tweet_text = """Just finished an amazing #NLP workshop! 🚀 
Check it out @ https://example.com/nlp-workshop 😊 
@AIResearcher you were right!"""

# Sample for MWE (Multi-Word Expression)
mwe_text = """New York is a beautiful city. 
Machine learning and natural language processing are subfields of artificial intelligence."""

print("Sample Text:")
print(sample_text)
print("\nTweet Text:")
print(tweet_text)
print("\nMWE Text:")
print(mwe_text)
```

## 3. Tokenization Techniques

### 3.1 Whitespace Tokenization

```python
whitespace_tokenizer = WhitespaceTokenizer()
whitespace_tokens = whitespace_tokenizer.tokenize(sample_text)

print("Whitespace Tokenization:")
print(f"Number of tokens: {len(whitespace_tokens)}")
print(f"Tokens: {whitespace_tokens}")
```

### 3.2 Punctuation-based Tokenization

```python
punct_tokenizer = WordPunctTokenizer()
punct_tokens = punct_tokenizer.tokenize(sample_text)

print("Punctuation-based Tokenization:")
print(f"Number of tokens: {len(punct_tokens)}")
print(f"Tokens: {punct_tokens}")
```

### 3.3 Treebank Tokenization

```python
treebank_tokenizer = TreebankWordTokenizer()
treebank_tokens = treebank_tokenizer.tokenize(sample_text)

print("Treebank Tokenization:")
print(f"Number of tokens: {len(treebank_tokens)}")
print(f"Tokens: {treebank_tokens}")
```

### 3.4 Tweet Tokenization

```python
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(tweet_text)

print("Tweet Tokenization:")
print(f"Number of tokens: {len(tweet_tokens)}")
print(f"Tokens: {tweet_tokens}")

# With different options
tweet_tokenizer_reduced = TweetTokenizer(reduce_len=True, strip_handles=False)
tweet_tokens_reduced = tweet_tokenizer_reduced.tokenize(tweet_text)

print("\nTweet Tokenization (with length reduction):")
print(f"Tokens: {tweet_tokens_reduced}")
```

### 3.5 MWE (Multi-Word Expression) Tokenization

```python
# First tokenize with word_tokenize
tokens = word_tokenize(mwe_text.lower())

# Create MWE tokenizer and add multi-word expressions
mwe_tokenizer = MWETokenizer()
mwe_tokenizer.add_mwe(('new', 'york'))
mwe_tokenizer.add_mwe(('machine', 'learning'))
mwe_tokenizer.add_mwe(('natural', 'language', 'processing'))
mwe_tokenizer.add_mwe(('artificial', 'intelligence'))

# Apply MWE tokenization
mwe_tokens = mwe_tokenizer.tokenize(tokens)

print("MWE Tokenization:")
print(f"Original tokens: {tokens}")
print(f"\nMWE tokens: {mwe_tokens}")
print(f"\nNotice how multi-word expressions are joined with underscores!")
```

## 4. Comparison of Tokenization Methods

```python
# Compare all tokenization methods
comparison_text = "It's a wonderful day! Let's learn NLP @ 2:30 PM."

comparison_data = {
    'Method': ['Whitespace', 'Punctuation-based', 'Treebank', 'Tweet'],
    'Tokens': [
        WhitespaceTokenizer().tokenize(comparison_text),
        WordPunctTokenizer().tokenize(comparison_text),
        TreebankWordTokenizer().tokenize(comparison_text),
        TweetTokenizer().tokenize(comparison_text)
    ]
}

df_comparison = pd.DataFrame(comparison_data)
df_comparison['Token Count'] = df_comparison['Tokens'].apply(len)

print(f"Original Text: '{comparison_text}'\n")
for idx, row in df_comparison.iterrows():
    print(f"{row['Method']} ({row['Token Count']} tokens):")
    print(f"  {row['Tokens']}\n")
```

## 5. Stemming Techniques

### 5.1 Porter Stemmer

```python
porter = PorterStemmer()

# Sample words for stemming
words_to_stem = [
    'running', 'runs', 'ran', 'runner',
    'easily', 'fairly', 'happily',
    'studies', 'studying', 'studied',
    'playing', 'plays', 'played', 'player',
    'connection', 'connections', 'connected', 'connecting'
]

print("Porter Stemmer Results:")
print(f"{'Word':<15} {'Stem':<15}")
print("-" * 30)
for word in words_to_stem:
    stem = porter.stem(word)
    print(f"{word:<15} {stem:<15}")
```

### 5.2 Snowball Stemmer

```python
snowball = SnowballStemmer('english')

print("Snowball Stemmer Results:")
print(f"{'Word':<15} {'Stem':<15}")
print("-" * 30)
for word in words_to_stem:
    stem = snowball.stem(word)
    print(f"{word:<15} {stem:<15}")
```

### 5.3 Comparison: Porter vs Snowball

```python
print("Comparison: Porter vs Snowball Stemmer")
print(f"{'Word':<20} {'Porter':<15} {'Snowball':<15} {'Same?'}")
print("-" * 60)

for word in words_to_stem:
    porter_stem = porter.stem(word)
    snowball_stem = snowball.stem(word)
    same = '✓' if porter_stem == snowball_stem else '✗'
    print(f"{word:<20} {porter_stem:<15} {snowball_stem:<15} {same}")
```

## 6. Lemmatization

### 6.1 WordNet Lemmatizer

```python
lemmatizer = WordNetLemmatizer()

# Words for lemmatization
words_to_lemmatize = [
    'running', 'runs', 'ran',
    'better', 'best', 'good',
    'feet', 'geese', 'mice',
    'studies', 'studying', 'studied',
    'playing', 'plays', 'played'
]

print("WordNet Lemmatizer Results (default - noun):")
print(f"{'Word':<15} {'Lemma':<15}")
print("-" * 30)
for word in words_to_lemmatize:
    lemma = lemmatizer.lemmatize(word)
    print(f"{word:<15} {lemma:<15}")
```

### 6.2 Lemmatization with POS Tags

```python
def get_wordnet_pos(treebank_tag):
    """Convert treebank POS tags to WordNet POS tags"""
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# Sample sentence
sentence = "The striped bats are hanging on their feet for best results"
tokens = word_tokenize(sentence)

# Get POS tags
pos_tags = pos_tag(tokens)

print("Lemmatization with POS Tags:")
print(f"{'Word':<15} {'POS':<10} {'Lemma (no POS)':<20} {'Lemma (with POS)':<20}")
print("-" * 75)

for word, pos in pos_tags:
    lemma_no_pos = lemmatizer.lemmatize(word.lower())
    wordnet_pos = get_wordnet_pos(pos)
    lemma_with_pos = lemmatizer.lemmatize(word.lower(), pos=wordnet_pos)
    print(f"{word:<15} {pos:<10} {lemma_no_pos:<20} {lemma_with_pos:<20}")
```

## 7. Stemming vs Lemmatization

```python
comparison_words = [
    'studies', 'studying', 'studied',
    'running', 'runs', 'ran',
    'better', 'best',
    'caring', 'cared', 'cares'
]

print("Stemming vs Lemmatization Comparison:")
print(f"{'Word':<15} {'Porter':<15} {'Snowball':<15} {'Lemma (verb)':<15}")
print("-" * 65)

for word in comparison_words:
    porter_result = porter.stem(word)
    snowball_result = snowball.stem(word)
    lemma_result = lemmatizer.lemmatize(word, pos=wordnet.VERB)
    print(f"{word:<15} {porter_result:<15} {snowball_result:<15} {lemma_result:<15}")

print("\nKey Observations:")
print("- Stemming: Crude chopping, may not be real words")
print("- Lemmatization: Returns actual dictionary words")
```

## 8. Complete Pipeline

```python
sample_pipeline_text = """The researchers are studying various Natural Language Processing techniques. 
They've been running multiple experiments with better algorithms."""

print("Original Text:")
print(sample_pipeline_text)
print("\n" + "="*80)

# Tokenization
print("\n1. TOKENIZATION (Treebank):")
tokens = TreebankWordTokenizer().tokenize(sample_pipeline_text)
print(tokens)

# Lowercasing
print("\n2. LOWERCASING:")
tokens_lower = [token.lower() for token in tokens]
print(tokens_lower)

# Stemming
print("\n3. STEMMING (Porter):")
stemmed_tokens = [porter.stem(token) for token in tokens_lower]
print(stemmed_tokens)

# Lemmatization
print("\n4. LEMMATIZATION (with POS):")
pos_tags = pos_tag(tokens)
lemmatized_tokens = [
    lemmatizer.lemmatize(word.lower(), pos=get_wordnet_pos(pos)) 
    for word, pos in pos_tags
]
print(lemmatized_tokens)
```
